```python
from astropy.io import fits
import h5py
import numpy as np
import bisect
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit
from scipy.signal import savgol_filter
from scipy.integrate import simps
from scipy import interpolate
import sys
import warnings
import glob
import os
from scipy.ndimage import gaussian_filter
import pysynphot as S
from astropy.table import Table

warnings.filterwarnings('ignore')

all_models = {}

# Directory with Dan's h5py models
model_dir = 'models/'

# THIS SHOULD BE SET TO YOUR PANDEIA DIRECTORY W/ JWST TRANSMISSION CURVES
pandeia = '/Users/ckilpatrick/scripts/pandeia/'

def get_bulla_model(file):

    # Check first three rows, should be Nobs(viewing angles), Nwave(wavelength bins),
    # Ntime(number of time bins)+time_start+time_final
    with open(file) as f:
        for i,line in enumerate(f):
            if i==0:
                nobs = int(line.strip())
            if i==1:
                nwave = int(line.strip())
            if i==2:
                ntime, t_start, t_end = line.split()
                ntime = int(ntime) ; t_start = float(t_start) ; t_end = float(t_end)
            if i>2:
                break

    # Construct a header for table
    header = ['wave']+['time_{0}'.format(str(i).zfill(3)) for i in np.arange(ntime)]

    # Construct a time array
    times = np.linspace(t_start, t_end, ntime)

    # Construct a cos(viewing angle) array
    angle = np.linspace(0, 1, nobs)

    t = Table.read(file, data_start=3,format='ascii',names=header)

    # Parse metadata from file name
    nph, mej, phi = os.path.basename(file).split('_')
    nph = float(nph.replace('nph',''))
    mej = float(mej.replace('mej',''))
    phi = float(phi.replace('phi','').replace('.txt',''))

    return(angle, times, header, nobs, nwave, ntime, mej, phi, t)

def get_bulla_lnu_nu(table, nobs_idx, header, nwave):
    # First get wavelength array from the first block of models
    wave = table[0:nwave]['wave'].data

    # Now group fluxes into a 2D array of time/wave
    flux = table[nobs_idx*nwave:nobs_idx*nwave+nwave]

    f = np.zeros((nwave, len(header[1:])))
    Lnu = np.zeros((len(header[1:]),nwave))
    for i,row in enumerate(header[1:]):
        f[:,i]=flux[row].data

    # Flux is in units of erg/s/cm2/Angstrom at 10pc, and we want to convert to
    # erg/s/Hz, so first multiply by 4*pi*(10pc)^2
    Llam = f * 4 * np.pi * (3.08568025e19)**2

    # Now flux is erg/s/Angstrom, so multiply by c/nu^2 = lambda^2/c to get to
    # erg/s/Hz.  c needs to be in units of angstrom * Hz
    for i in np.arange(len(header[1:])):
        Lnu[i,:] = Llam[:,i] * wave**2 / (2.998e18)

    # Finally convert wave to nu
    nu = 2.998e18 / wave

    return(Lnu, nu)

def blackbody(wave, a, t):
    return(a/wave**5 * 1/(np.exp(143843215.0/(wave * t))-1.0))

# Testing integration for arbitrary blackbody SEDs
wave = np.linspace(100, 500000, 500000)
for temp in [3000,5000,7000,10000,12000]:
    bb = blackbody(wave, 1.0, temp)
    value1 = simps(wave, bb)
    value2 = simps(wave, bb*wave)
    print(value2/value1*temp)

# Modified blackbody based on recent model from Dan
def mod_blackbody(wave, a, t, c, micron=True):
    if micron: 
        expwave = wave * 1.0e4
    else:
        expwave = wave
    return(a/wave**5 * 1/(np.exp(143843215/(expwave * t))-1.0)*np.exp(-c*expwave**1.31555381e-01))

# Will scale a light curve after we define lum_func in next cell
def scaled_lum(t, a, b):
    c = t/a
    return(b*10**lum_func(c))

# Scale a blackbody SED to match model at a specific wavelength (for splicing SEDs together)
def calc_scale(luminosity, scale_wave, temp):
    normbb = blackbody(scale_wave, 1.0, temp)
    scale = luminosity/normbb
    return(scale)

# Estimate the effective temperature of an SED assuming blackbody
def estimate_temperature(wave, luminosity):
    w = np.array(wave).astype('double')
    l = np.array(luminosity).astype('double')
    value1 = simps(w, l)
    value2 = simps(w, l*w)
    print('v1:',value1)
    print('v2:',value2)
    temp = 53146846.449553266 * value1/value2
    return(temp)

# Generic power law function for light curve
def power_law(time, a, b):
    return(a*time**b)

# Get all filters from Pandeia directory (see pandeia variable above)
def get_filters(inst='nircam'):
    if inst.lower()=='nircam':
        directory = pandeia+'jwst/nircam/filters/'
        filts = glob.glob(directory+'*w_trans*')
        filts += glob.glob(directory+'*w2_trans*')
    elif inst.lower()=='miri':
        directory= pandeia+'jwst/miri/filters/'
        filts = glob.glob(directory+'*w_trans*')
    elif inst.lower()=='wfc3':
        directory='/Users/ckilpatrick/scripts/cdbs/comp/wfc3/'
        filts = glob.glob(directory+'wfc3uvis2_f*w_mjd_006_syn.fits')
        filts += glob.glob(directory+'wfc3uvis2_f*lp_mjd_006_syn.fits')
        filts += glob.glob(directory+'wfc3_ir_f*w_004_syn.fits')        
    else:
        return([])
        
    outfilts = []
    for filt in filts:
        base = os.path.split(filt)[1]
        if inst!='wfc3' or 'wfc3_ir' in base:
            name = inst.lower()+'_'+base.split('_')[2]
        else:
            name = inst.lower()+'_'+base.split('_')[1]
        hdu = fits.open(filt)
        if inst!='wfc3': 
            wave = np.array([float(el[0])*1.0e4 for el in hdu[1].data])
        else:
            wave = np.array([float(el[0]) for el in hdu[1].data])
        tran = np.array([float(el[1]) for el in hdu[1].data])
        
        # Only consider wavelengths < 30 microns
        mask = wave < 30.0*1.0e4
        wave = wave[mask]
        tran = tran[mask]
        
        bp = S.ArrayBandpass(wave, tran, name=name, waveunits='angstrom')
        outfilts.append(bp)
    
    return(outfilts)

def get_ground_filts():
    names=['sdss,g','sdss,r','sdss,i','sdss,z',
          'johnson,u','johnson,b','johnson,v','johnson,r','johnson,i',
          'bessell,j','bessell,h','bessell,k']
    
    bps=[]
    for n in names:
        bp = S.ObsBandpass(n)
        bp.name = n
        bps.append(bp)
    
    return(bps)

def rebin_spec(wave, specin, wavnew):
    spec = S.ArraySpectrum(wave=wave, flux=specin)
    f = np.ones(len(wave))
    filt = S.ArrayBandpass(wave, f, waveunits='angstrom')
    obs = S.Observation(spec, filt, binset=wavnew, force='taper')
 
    return obs.binflux
```

```python
###
### Below defining models to output light curves
###
nnu = [] ; ttime = [] ; LLnu_all = [] ; nname = []

#name = 'knova_d1_n10_m0.050_vk0.10_fd1.0_Xlan1e-1.0.h5'
#fin    = h5py.File(model_dir+name,'r')
#nu     = np.array(fin['nu'],dtype='d')
#times  = np.array(fin['time'])
#times  = times/3600.0/24.0
#Lnu_all   = np.array(fin['Lnu'],dtype='d')

#nnu.append(nu) ; ttime.append(times) ; LLnu_all.append(Lnu_all) ; nname.append('knova_m0.050_vk0.10_Xlan1e-1.0')

# First model is what we used for GW170817 - basic vanilla model
name = 'knova_d1_n10_m0.035_vk0.15_Xlan1e-2.0.h5'
fin    = h5py.File(model_dir+name,'r')
nu     = np.array(fin['nu'],dtype='d')
times  = np.array(fin['time'])
times  = times/3600.0/24.0
Lnu_all   = np.array(fin['Lnu'],dtype='d')

name2  = 'knova_d1_n50_m0.025_vk0.25_fd0.3_Xlan1e-6.0_vs0.32_ns12.0.h5'
fin2   = h5py.File(model_dir+name2,'r')
nu2    = np.array(fin2['nu'],dtype='d')
times2 = np.array(fin2['time'])
times2 = times2/3600.0/24.0

Lnu_all2 = np.array(fin2['Lnu'],dtype='d')
f = interpolate.interp2d(nu2, times2, Lnu_all2, kind='linear')
Lnu_all2_interp = f(nu, times)
Lnu_all = Lnu_all + Lnu_all2_interp

nnu.append(nu) ; ttime.append(times) ; LLnu_all.append(Lnu_all) ; nname.append('gw170817')

# Second model is a red kilonova with larger ejecta mass
name = 'knova_d1_n10_m0.050_vk0.15_Xlan1e-2.0.h5'
fin    = h5py.File(model_dir+name,'r')
nu     = np.array(fin['nu'],dtype='d')
times  = np.array(fin['time'])
times  = times/3600.0/24.0
Lnu_all   = np.array(fin['Lnu'],dtype='d')

nnu.append(nu) ; ttime.append(times) ; LLnu_all.append(Lnu_all) ; nname.append('red_m0.050')

# Second model is a red kilonova with larger ejecta mass
name = 'knova_d1_n10_m0.010_vk0.10_fd1.0_Xlan1e-2.0.h5'
fin    = h5py.File(model_dir+name,'r')
nu     = np.array(fin['nu'],dtype='d')
times  = np.array(fin['time'])
times  = times/3600.0/24.0
Lnu_all   = np.array(fin['Lnu'],dtype='d')

nnu.append(nu) ; ttime.append(times) ; LLnu_all.append(Lnu_all) ; nname.append('red_m0.010')

# Third model is a red kilonova with lower ejecta mass
name = 'knova_d1_n10_m0.001_vk0.10_fd1.0_Xlan1e-2.0.h5'
fin    = h5py.File(model_dir+name,'r')
nu     = np.array(fin['nu'],dtype='d')
times  = np.array(fin['time'])
times  = times/3600.0/24.0
Lnu_all   = np.array(fin['Lnu'],dtype='d')

nnu.append(nu) ; ttime.append(times) ; LLnu_all.append(Lnu_all) ; nname.append('red_m0.001')

# Third model is a red kilonova with lower ejecta mass
name = 'knova_d1_n10_m0.1_vk0.10_fd1.0_Xlan1e-2.0.h5'
fin    = h5py.File(model_dir+name,'r')
nu     = np.array(fin['nu'],dtype='d')
times  = np.array(fin['time'])
times  = times/3600.0/24.0
Lnu_all   = np.array(fin['Lnu'],dtype='d')

nnu.append(nu) ; ttime.append(times) ; LLnu_all.append(Lnu_all) ; nname.append('red_m0.100')

# Third model is a red kilonova with lower ejecta mass
name = 'knova_d1_n10_m0.030_vk0.10_Xlan1e-2.0.h5'
fin    = h5py.File(model_dir+name,'r')
nu     = np.array(fin['nu'],dtype='d')
times  = np.array(fin['time'])
times  = times/3600.0/24.0
Lnu_all   = np.array(fin['Lnu'],dtype='d')

nnu.append(nu) ; ttime.append(times) ; LLnu_all.append(Lnu_all) ; nname.append('red_m0.030')

# Third model is a red kilonova with lower ejecta mass
name = 'knova_d1_n10_m0.040_vk0.30_Xlan1e-4.0.h5'
fin    = h5py.File(model_dir+name,'r')
nu     = np.array(fin['nu'],dtype='d')
times  = np.array(fin['time'])
times  = times/3600.0/24.0
Lnu_all   = np.array(fin['Lnu'],dtype='d')

nnu.append(nu) ; ttime.append(times) ; LLnu_all.append(Lnu_all) ; nname.append('blue_m0.040')

# First model is what we used for GW170817 - basic vanilla model
name = 'knova_d1_n10_m0.035_vk0.15_Xlan1e-2.0.h5'
fin    = h5py.File(model_dir+name,'r')
nu     = np.array(fin['nu'],dtype='d')
times  = np.array(fin['time'])
times  = times/3600.0/24.0
Lnu_all   = np.array(fin['Lnu'],dtype='d')

name2  = 'knova_d1_n50_m0.025_vk0.25_fd0.3_Xlan1e-6.0_vs0.32_ns12.0.h5'
fin2   = h5py.File(model_dir+name2,'r')
nu2    = np.array(fin2['nu'],dtype='d')
times2 = np.array(fin2['time'])
times2 = times2/3600.0/24.0

Lnu_all2 = np.array(fin2['Lnu'],dtype='d')
f = interpolate.interp2d(nu2, times2, Lnu_all2, kind='linear')
Lnu_all2_interp = f(nu, times)
Lnu_all = Lnu_all + Lnu_all2_interp

nnu.append(nu) ; ttime.append(times) ; LLnu_all.append(Lnu_all) ; nname.append('gw170817_boosted')

#nnu = [] ; ttime = [] ; LLnu_all = [] ; nname = []

#for file in glob.glob('models/knova_d1_n10_m0.03*vk0.1*fd1.0*.h5'):
#    fin = h5py.File(file, 'r')
#    nu     = np.array(fin['nu'],dtype='d')
#    times  = np.array(fin['time'])
#    times  = times/3600.0/24.0
#    Lnu_all   = np.array(fin['Lnu'],dtype='d')
    
#    name = os.path.split(file)[1].replace('.h5','')
    
#    nnu.append(nu) ; ttime.append(times) ; LLnu_all.append(Lnu_all) ; nname.append(name)

# Get all of the JWST filts we want to output
#filts = get_filters(inst='nircam') + get_filters(inst='miri')+get_filters(inst='wfc3')
filts = get_filters(inst='wfc3')+get_ground_filts()

# This function takes one of the sets of times, nu, Lnu_all from Dan's models and extrapolates both
# SED and light curve to 30 micron and +150 days, respectively.  Then output JWST light curves
def generate_full_jwst_model(times, nu, Lnu_all, name=None, boosted=False, out=True, time_steps=250, 
                             wave_steps=10000):

    # for example, get the spectrum at day 2.0
    sed_fits = {'time':[], 'temp':[], 'fit': [ 6.01816380e+03, -2.94812545e-01], 'lum': [], 'full_sed':[], 'lam': None}
    for it,t in enumerate(times):
        if t < 0: continue
        if t > 20: break
        #continue
        Lnu = Lnu_all[it,:]

        c    = 2.99e10
        lam  = c/nu*1e8
        Llam = Lnu*nu**2.0/c/1e8

        idx = int(len(lam)/2)
        dv = (lam[idx]-lam[idx+1])/lam[idx]
        ngauss = int(0.03 / dv)

        mask = Llam<=0.0
        try:
            f = interpolate.interp1d(lam[~mask], Llam[~mask], kind='slinear', bounds_error=False,
                                    fill_value='extrapolate')
        except:
            break
        Llam = f(lam)
        Llam = gaussian_filter(Llam, sigma=ngauss)

        # Scale to 9.5 microns
        #extrap_start = 95000.0
        #extrap_end   = 300000.0
        #idx = np.argmin(np.abs(lam-extrap_start))
        #if lam[idx] > extrap_start: idx = idx-1
        #mask = lam < extrap_start
        #mask2 = (lam > 4000.0) & (lam < extrap_start)
        # Get estimate of blackbody temperature from spectrum
        #temp = estimate_temperature(lam[mask2], Llam[mask2])

        #if len(sed_fits['time'])>40:
        #    popt, pcov = curve_fit(power_law,sed_fits['time'], sed_fits['temp'], p0=sed_fits['fit'])
        #    
        #    if np.abs((power_law(t, *popt)-temp)/power_law(t, *popt))<0.3:
        #        sed_fits['time'].append(t)
        #        sed_fits['temp'].append(temp)
        #        sed_fits['fit']=popt
        #    else:
        #        temp = power_law(t, *popt)
        #        sed_fits['time'].append(t)
        #        sed_fits['temp'].append(temp)
        #else:
        #    sed_fits['time'].append(t)
        #    sed_fits['temp'].append(temp)

        #scale = calc_scale(Llam[idx], lam[idx], temp)

        #tlam=np.linspace(extrap_start,extrap_end,1500)
        #tLum=blackbody(tlam, scale, temp)
        #Llam = np.concatenate((np.flip(Llam[mask]), tLum), axis=0)
        #print(Llam)
        #try:
        Llam = savgol_filter(Llam, 11, 3)
        #except:
        #    pass
        #lam = np.concatenate((np.flip(lam[mask]), tlam), axis=0)

        total_lum=simps(Llam, lam)
        sed_fits['lum'].append(total_lum)
        sed_fits['full_sed'].append(Llam)
        sed_fits['time'].append(t)
        sed_fits['lam'] = lam

    # Now we need to extrapolate the light curve to t_max using the generic LC shape
    for key in sed_fits.keys(): sed_fits[key] = np.array(sed_fits[key])
    mask = (sed_fits['time'] > 5.0) & ~np.isnan(sed_fits['lum']) & ~np.isinf(sed_fits['lum'])
    popt, pcov = curve_fit(scaled_lum, sed_fits['time'][mask], sed_fits['lum'][mask], p0=(1.0, 1.0))
    
    test_lum = scaled_lum(sed_fits['time'], *popt)
    scaled = np.array(np.abs((test_lum - sed_fits['lum'])/test_lum))
    mask2 = (sed_fits['time'] > 5.0) & (scaled > 0.25)
    
    if len(sed_fits['time'][mask2])>5:
        break_idx = np.min(np.where(mask2))
        last_time = sed_fits['time'][break_idx-1]
        last_lum = sed_fits['lum'][break_idx-1]
        last_sed = sed_fits['full_sed'][break_idx-1]
        
        sed_fits['time'] = sed_fits['time'][0:break_idx-1]
        sed_fits['lum'] = sed_fits['lum'][0:break_idx-1]
        sed_fits['full_sed'] = sed_fits['full_sed'][0:break_idx-1]
    else:
        last_time = sed_fits['time'][-1]
        last_lum = sed_fits['lum'][-1]
        last_sed = sed_fits['full_sed'][-1]
           
    t_max = 150.0
    extrap_time = np.linspace(last_time, t_max, 200)

    # Now freeze spectrum in place from the last data point and extrapolate to t_max
    t_max = 150.0
    for key in sed_fits.keys(): sed_fits[key] = list(sed_fits[key])
    for time in np.linspace(last_time, t_max, 500):
        lum = scaled_lum(time, *popt)
        Llam = lum/last_lum * last_sed
        sed_fits['full_sed'].append(Llam)
        sed_fits['lum'].append(lum)
        sed_fits['time'].append(time)
     
    # This part of the function adds additional luminosity for an arbitrary
    # r-process species given atomic mass, energy per decay, half life, thermalization
    if boosted:
        for key in sed_fits.keys(): sed_fits[key] = np.array(sed_fits[key])
        # Only add energy at late times to account for the fact that we're assuming
        # the energy is instantaneously injected so must be optically thin
        mask_boost = sed_fits['time'] > 0.0
        # Get fiducial luminosity for 1.0e-5 Msun of Californium-254
        mass        = 1.0e-5       # Total mass of material
        atomic_mass = 252.647831   # atomic mass (for Cf-254)
        Q           = 0.000296403  # energy per decay in ergs (185 MeV for Cf-254)
        t2          = 60.5         # Half-life in days
        therm       = 0.3          # Thermalization coefficient
        
        # Fiducial luminosity by which the light curve is boosted due to this species
        Lboost = therm * Q * mass * 1.989e33 /(atomic_mass*1.67e-24)/(t2 * 86400)
        
        # Now luminosity will decline as 2.0**(-t/60.5 days)
        Lboost = Lboost * 2.0**(-sed_fits['time'][mask_boost]/t2)
        
        # Now add this luminosity to the light curve and SED
        for j,idx in enumerate(np.where(mask_boost)[0]):
            sed_fits['full_sed'][idx,:] = (1.0+Lboost[j]/sed_fits['lum'][idx])*sed_fits['full_sed'][idx,:]
        sed_fits['lum'][mask_boost] += Lboost
        
        for i in np.arange(len(sed_fits['time'])):
            sed_fits['lum'][i]=simps(sed_fits['full_sed'][i,:], sed_fits['lam'])
        

    #plt.plot(sed_fits['time'], sed_fits['lum'], label=name)
    #plt.yscale('log')
        
    newtime = np.linspace(np.min(sed_fits['time']), np.max(sed_fits['time']), time_steps)
    newwave = np.linspace(np.min(sed_fits['lam']), np.max(sed_fits['lam']), wave_steps)
        
    dummy_row = [[0.0]]
    header = ['rest_time']
    for bp in filts: 
        dummy_row.append([0.0])
        header.append(bp.name)
        
    table = Table(dummy_row, names=header).copy()[:0]
    
    sed_fits['full_sed'] = np.array(sed_fits['full_sed'])
    sed_fits['time'] = np.array(sed_fits['time'])
    sed_fits['lam'] = np.array(sed_fits['lam'])
    
    new_sed = np.zeros((len(newtime), len(newwave)))
    
    for i,time in enumerate(newtime):
        
        idx1 = np.argmin(np.abs(sed_fits['time']-time))
        
        fluxscale = 1./(4*np.pi * (10.0 * 3.08568025e18)**2)
        fluxval = fluxscale * sed_fits['full_sed'][idx1,:]
        mask = fluxval < 0.
        fluxval[mask] = 0.0
                
        spectrum = S.ArraySpectrum(wave=sed_fits['lam'], flux=fluxval, 
                                   waveunits='angstrom', fluxunits='flam')
        
        spbinflux = rebin_spec(spectrum.wave, spectrum.flux, newwave)
        new_sed[i,:]=spbinflux
                
        row = [time]
        
        for bp in filts:
            try:
                obs = S.Observation(spectrum, bp, binset=newwave, force='taper')
                row.append(obs.effstim('abmag'))
            except:
                row.append(np.nan)
        
        table.add_row(row)
    
    with open('seds/'+name+'.pkl', 'wb') as f:
        wave = newwave
        time = newtime
        sed = new_sed
        np.save(f, wave)
        np.save(f, time)
        np.save(f, sed)
        
    # Format to reasonable floating point
    for key in table.keys():
        newrow = ['%7.4f'%row[key] for row in table]
        table[key] = table[key].astype(str)
        table[key] = np.array(newrow)
        table[key] = table[key].astype(float)
        
    for key in table.keys():
        if key=='rest_time': continue
        
        plt.plot(table['rest_time'], table[key])
    
    plt.xlim([0,150])
    plt.ylim([-5,-16])
    
    table.meta['name'] = name
    table.write('seds/'+name+'.dat', format='ascii.ecsv')
    
    return(None)

#for filename in glob.glob('bulla/*.txt'):
#    angle, times, header, nobs, nwave, ntime, mej, phi, t = get_bulla_model(filename)
#    for i in np.arange(nobs):
#        Lnu, nu = get_bulla_lnu_nu(t, i, header, nwave)
#        
#        name = 'bulla/mej{0}_phi{1}_cosangle{2}.dat'.format(str('%7.4f'%mej).strip(),str('%7.2f'%phi).strip(),
#                                                      '%1.4f'%angle[i])
#        
#        generate_full_jwst_model(times, nu, Lnu, boosted=False, name=name)

for nu,times,Lnu_all,name in zip(nnu,ttime,LLnu_all,nname):
    #if 'gw170817' not in name: continue
    boosted=False
    if 'boosted' in name: boosted=True
    generate_full_jwst_model(times, nu, Lnu_all, boosted=boosted, name=name)
    
#plt.legend()
```
